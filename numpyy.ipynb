```python
#Quarteriy Difference
import numpy as np
s_x=np.array([[2,5,6,5],[4,8,6,5]])
s_y=np.array([[6,7,5,9],[7,5,6,4]])
quarterly_difference=s_y-s_x
print(quarterly_difference)
```

```python
#Quarters having literacy rate less than 2
less_than_2=s_x<2
print(less_than_2)
```

```python
#cummulative sum of literacy rates by the end of ever quarter each year and finding quarters with  a sum greater or equal to 8: 
x_cumsum=np.cumsum(s_x,axis=1)
print(x_cumsum)
print(x_cumsum>=8)
y_cumsum=np.cumsum(s_y,axis=1)
print(y_cumsum)
print(y_cumsum>=8)
```

```python
##mean of each years literacy rate of stste X and mean of each quarterlys literacy rate of state Y:
yearly_mean_x=np.mean(s_x,axis=1)
print(yearly_mean_x)
```

```python
quarterly_mean_y=np.mean(s_y,axis=0)
print(quarterly_mean_y)
```

```python
##use map, filter,and lambda Functions with numpy
#map
numbers=[1,2,3,4,5]
squared_numbers=map(lambda x: x**2,numbers)
print(list(squared_numbers))
```

```python
#Filter:
import numpy as np
numbers=[1,2,3,4,5]
even_numbers = filter(lambda x: x%2==0,numbers)
print(list(even_numbers))
```

```python
#Reduce:
from functools import reduce
numbers=[1,2,3,4,5]
sum_of_numbers=reduce(lambda x,y:x+y,numbers)
print(sum_of_numbers)
```
