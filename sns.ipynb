```python
import pandas as pd
df=pd.read_csv('dataset/auto_mpg.csv')
df
```

```python
#1.BOX PLOT:
import seaborn as sns
import matplotlib.pyplot as plt
sns.boxplot(data=df,y=df['mpg'])
plt.show()
```

```python
#2.SCATTER PLOT:
sns.scatterplot(data=df,x=df['mpg'],y=df['horsepower'])
plt.show()
```

```python
#3.BAR PLOT:
sns.barplot(data=df,x=df['mpg'],y=df['horsepower'])
plt.show()
```

```python
sns.barplot(data=df,x=df['origin'],y=df['model_year'])
plt.show()
```

```python
#4.HISTOGRAM:
sns.histplot(data=df,x=df['horsepower'])
plt.show()
```

```python
sns.histplot(data=df,x=df['horsepower'],bins=20,hue=True,kde=True)
plt.show()
```

```python
#5.LINE PLOT:
sns.lineplot(data=df,x=df['mpg'],y=df['horsepower'])
plt.show()
```

```python
#6.COUNT PLOT:
sns.countplot(data=df,x=df['mpg'])
plt.show()
```

```python
#7.HEATMAP:
sns.heatmap(data=df[['mpg']])
plt.show()
```

```python
sns.heatmap(data=df[['mpg','horsepower']])
plt.show()
```
